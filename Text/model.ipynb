```python
# Library imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
# Reading the CSV file
raw_df = pd.read_csv("Datasets/Sym-DiseaseTraining.csv")
raw_df.info()
```

```python
raw_df.describe()
```

```python
#Dividing the input and output datas
input_data = raw_df.iloc[:,:-1]
output_data = raw_df.iloc[:,-1]

print(input_data)
```

```python
print(output_data)
```

```python
# Dividing the Validation Data
itrain,ivar, otrain, ovar = train_test_split(input_data, output_data, test_size=0.1,random_state=30)
```

```python
print(itrain.shape)
print(otrain.shape)
print(ivar.shape)
print(ovar.shape)
```

```python
Text_model = RandomForestClassifier(n_jobs=-1, random_state=30)
%time
Text_model.fit(itrain,otrain)
```

```python
Text_model.score(itrain,otrain)
```

```python
Text_model.predict(ivar[2:4])
```

```python
ovar[2:4]
```

```python
test_data = pd.read_csv("Datasets/Sym-DiseaseTesting.csv")
test_input= test_data.iloc[:,:-1]
test_output= test_data.iloc[:,-1]
output = Text_model.predict(test_input)
print(output)
```

```python
Text_model.score(test_input,test_output)
```
